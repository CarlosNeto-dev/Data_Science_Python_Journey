**‚úèÔ∏è Como funciona a fun√ß√£o 'lambda' nos c√≥digos de Python? ‚úèÔ∏è**
- A fun√ß√£o lambda √© uma fun√ß√£o anonima, ou seja, ela 'esconde' a fun√ß√£o, outrossim, √© mais simplificada e simples.

```python
random_value = 1000
double_of_value = lambda x: x * 2
double_of_value(random_value)
```

**_Vou fazer a mesma coisa, por√©m sem utilizar a fun√ß√£o lambda._**

```python
random_value = 1000
def double_of_value(value):
    return value * 2


double_of_value(random_value)
```

**_Percebe-se que a fun√ß√£o lambda simplifica a codifica√ß√£o, outrossim, √© bastante utilizado em v√°rios programas. Por√©m, √© recomendado usar lambda para fun√ß√µes mais simples._**

## ‚ú® UTILIZA√á√ÉO DO LAMBDA JUNTO COM A FUN√á√ÉO MAP() ‚ú®

```python
list_of_values = [1000, 1500, 100, 350]
double_of_value = list(map(lambda x: x * 2, list_of_values))
print(double_of_value)
```

**_Nota-se, que consegui simplificar o meu c√≥digo mais ainda com a utiliza√ß√£o da fun√ß√£o "map()" e "lambda". Sendo assim, um recurso bastante √∫til e que n√£o pode ser descart√°vel._**

## ‚ú® UTILIZANDO O LAMBDA JUNTO COM A A FUN√á√ÉO SORTED() E KEY() ‚ú®

```python
students = [
    {"NAME": "Carlos", "AGE": "18", "SCORE": 9},
    {"NAME": "Pedro", "AGE": "20", "SCORE": 9.5},
    {"NAME": "Jo√£o", "AGE": "32", "SCORE": 7.2}
]

ordened_by_score = sorted(students, key=lambda student: student["SCORE"], reverse=True)

for c,s in enumerate(ordened_by_score, start=1):
    print(f"{c} - {s} ")
```

**_Como pode notar, utilizamos o "sorted()" com o prop√≥sito de ordenar um determinado elemento, especificamos o elemento quando
utilizamos o "key()" o qual espera receber fun√ß√µes. Com isso, podemos usar o lambda j√° que √© uma fun√ß√£o "an√¥nima". Logo, a
fun√ß√£o sorted(recebe_uma_lista/tupla/dicion√°rio, key=fun√ß√£o_aleat√≥ria). Em detrimento disso, quando usamos "lambda student:
student["SCORE"]", espera-se receber os valores de "SCORE" nos dicion√°rios, assim, ordenando-os. Outrossim, o "reverse=True
" ordena em ordem decrescente._**

**_Vou fazer a mesma coisa, por√©m sem utilizar a fun√ß√£o lambda._**

```python
students = [
    {"NAME": "Carlos", "AGE": "18", "SCORE": 9},
    {"NAME": "Pedro", "AGE": "20", "SCORE": 9.5},
    {"NAME": "Jo√£o", "AGE": "32", "SCORE": 7.2}
]

def the_highest_scores_of_students(student):
    return student["SCORE"]


ordened_by_score = sorted(students, key=the_highest_scores_of_students, reverse=True)

for c,s in enumerate(ordened_by_score, start=1):
    print(f"{c} - {s} ")
```

**_Percebe-se, usando a fun√ß√£o lambda, consegui otimizar mais a minha codifica√ß√£o. Usar lambda n√£o √© uma regra que deve ser seguida a todo custo. Por√©m, √© uma boa alternativa para deixar o c√≥digo
mais simples e otimizado._**
