# FUNÇÃO LAMBDA NO PYTHON
**Como funciona a função 'lambda' nos códigos de Python?**
- A função lambda é uma função anônima, ou seja, ela 'esconde' a função, outrossim, é mais simplificada e simples.

```python
random_value = 1000
double_of_value = lambda x: x * 2
double_of_value(random_value)
```

**_Vou fazer a mesma coisa, porém sem utilizar a função lambda._**

```python
random_value = 1000
def double_of_value(value):
    return value * 2


double_of_value(random_value)
```

**_Percebe-se que a função lambda simplifica a codificação, outrossim, é bastante utilizado em vários programas. Porém, é recomendado usar lambda para funções mais simples._**

## UTILIZAÇÃO DO LAMBDA JUNTO COM A FUNÇÃO MAP()

```python
list_of_values = [1000, 1500, 100, 350]
double_of_value = list(map(lambda x: x * 2, list_of_values))
print(double_of_value)
```

**_Nota-se, que consegui simplificar o meu código mais ainda com a utilização da função "map()" e "lambda". Sendo assim, um recurso bastante útil e que não pode ser descartável._**
