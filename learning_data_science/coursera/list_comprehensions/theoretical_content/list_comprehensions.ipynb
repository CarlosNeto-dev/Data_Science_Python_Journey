# LIST COMPREHENSIONS
**_O quão importante é utilizar o list comprehensions durante a codificação?_**
- Esse conteúdo é muito interessante e pertinente na codificação rotineira. Pois, consegue simplificar e deixar mais otimizado.

**_Usando com list comprehensions_**

```python
list_of_values = [1000, 100, 50, 20, 400]

double_of_value = [v * 2 for v in list_of_values]
print(double_of_value)
```

**_Usando sem list comprehensions_**

```python
list_of_values = [1000, 100, 50, 20, 400]

double_of_value = []

for v in list_of_values:
    double_of_value.append(v * 2)
print(double_of_value)
```

**_Percebe-se que quando eu uso a utilidade list comprehensions, meu código ficar consideravelmente menor e otimizado. Consigo colocar a multiplicação para cada valor em uma lista em apenas uma linha de código._**
