## ‚ùì **_Qual o prop√≥sito da utiliza√ß√£o da fun√ß√£o "map()" na codifica√ß√£o?_** ‚ùì

*Usando a fun√ß√£o "map()" na codifica√ß√£o:*

```python
list_of_prices = [100, 1000, 4250, 3020]

def square_values(value):
    return value ** 2


square_values_of_list = list(map(square_values, list_of_prices))
print(square_values_of_list)
```

*Sem usar a fun√ß√£o "map()" na codifica√ß√£o:*

```python
list_of_prices = [100, 1000, 4250, 3020]
square_values_of_list = []

def square_values(value):
    return value ** 2


for v in list_of_prices:
    square_values_of_list.append(square_values(v))

print(square_values_of_list)
```

**Observa-se que sem a utiliza√ß√£o da fun√ß√£o "map()" o c√≥digo ficou muito mais longo. Ou seja, sem o "map()" tive que criar uma
lista e usei o ".append()" para acrescentar esses valores para a "square_values_of_list". Em detrimento disso, usar a
fun√ß√£o "map()" facilita e otimiza a codifica√ß√£o.**

## **_üìö Como funciona a fun√ß√£o "map()"? üìö_**

**A fun√ß√£o "map()" necessita da seguinte estrutura:**

*_map(any_function, list/dict/tuple)_* --> **_Logo, √© necess√°rio passar uma fun√ß√£o como primeiro par√¢metro e logo em seguida, passar uma tupla, lista ou dicion√°rio como segundo par√¢metro._**

**Explica√ß√£o: O map() vai pegar os primeiro valor de uma tupla, lista ou dicion√°rio e vai aplicar a fun√ß√£o que voc√™ colocou no primeiro par√¢metro nela, sendo assim, repete-se esse ciclo at√©
acabar os elementos.**

## **_üßæ Lazy Iterator em map()üßæ_**

```python
list_of_prices = [100, 1000, 4250, 3020]

def square_values(value):
    return value ** 2


square_numbers = map(square_values, list_of_prices)
print(square_numbers)
```

**Quando n√£o usamos o "list()" antes da fun√ß√£o map(), ele retorna esse valor iter√°vel. O map() n√£o cria uma lista imediatamente,
logo, esse valor iter√°vel √© apenas uma receita, sendo assim, sabe gerar os resultados solicitados. Por√©m, para gerar esses valores,
√© necess√°rio solicitar. Contudo, para consumir esse objeto √© necess√°rio que voc√™ use alguns m√©todos...**

*Primeiro M√©todo:*

```python
list_of_prices = [100, 1000, 4250, 3020]

def square_values(value):
    return value ** 2


square_numbers = map(square_values, list_of_prices)
print(next(square_numbers))
print(next(square_numbers))
print(next(square_numbers))
print(next(square_numbers))
```

**Com esse primeiro m√©todo, consume-se o valor iter√°vel pela fun√ß√£o "next()" que vai entregar o valor na medida que voc√™ solicita.**

*Segundo M√©todo:*

```python
list_of_prices = [100, 1000, 4250, 3020]

def square_values(value):
    return value ** 2


square_numbers = map(square_values, list_of_prices)
print(list(square_numbers))
```

**Com esse segundo m√©todo, consume-se o valor iter√°vel pelo "list()" junto com "print()". Logo, n√£o cria uma lista de imediato.**

*Varia√ß√£o do Segundo M√©todo:*

```python
list_of_prices = [100, 1000, 4250, 3020]

def square_values(value):
    return value ** 2


square_numbers = list(map(square_values, list_of_prices))
print(square_numbers)
```

**J√° nessa varia√ß√£o do segundo m√©todo, o valor iter√°vel j√° √© consumido de imediato pelo "list()". Portanto, quando vai realizar o "print(square_numbers)" j√° n√£o √© mais um valor iter√°vel.**

*Terceiro M√©todo:*

```python
list_of_prices = [100, 1000, 4250, 3020]

def square_values(value):
    return value ** 2


square_numbers = map(square_values, list_of_prices)
for iterator_value in square_numbers:
    print(iterator_value)
```

**Nesse terceiro m√©todo, cria-se uma estrutura de repeti√ß√£o "for" que vai consumindo o valor iter√°vel a cada looping. Ou seja, de forma gradativa e n√£o imediata.**

**üü• Para qualquer m√©todo, se eu tentar consumir o valor iter√°vel mais que o necess√°rio, o que vai acontecer com o c√≥digo? üü•**

```python
list_of_prices = [100, 1000, 4250, 3020]

def square_values(value):
    return value ** 2


square_numbers = map(square_values, list_of_prices)
print(next(square_numbers))
print(next(square_numbers))
print(next(square_numbers))
print(next(square_numbers))
print(next(square_numbers))
```

**Diante disso que acabamos de observar, quando tentamos ultrapassar a quantidade de valores iter√°veis existentes dentro do "map()", ir√° d√° um erro de "StopIteration", o qual n√£o vai ter mais valores
para achar.**

## **_üõ†Ô∏è FORMAS DE SIMPLIFICA√á√ÉO DA FUN√á√ÉO "MAP()"üõ†Ô∏è_**

*Usando o "map()" com a fun√ß√£o an√¥nima "lambda":*

```python
list_of_prices = [100, 1000, 4250, 3020]

square_numbers = list(map(lambda value: value ** 2, list_of_prices))
print(square_numbers)
```

**Usar com a fun√ß√£o "lambda" pode ser uma √≥tima alternativa para otimizar a codifica√ß√£o, j√° que n√£o √© necess√°rio v√°rias linhas de c√≥digo para criar uma fun√ß√£o relativamente simples.**

*Usando o "map()" com a fun√ß√£o "lambda" e "filter()":*

```python
list_of_prices = [100, 1000, 4250, 3020]

filtered_numbers = list(
    map(
        lambda filtered_value: filtered_value ** 2,
        filter(lambda value: value > 2000, list_of_prices)
    )
)
print(filtered_numbers)
```

**Usei a fun√ß√£o "filter()" para filtrar os valores da lista "list_of_prices" que s√£o maiores que 2000. Com os valores j√° filtrados, a fun√ß√£o "map()" vai deix√°r-los ao quadrado.**

**A fun√ß√£o "filter()" funciona de forma parecida com a fun√ß√£o "map()", ela espera receber uma fun√ß√£o aleat√≥ria e alguma lista/tupla/dicion√°rio. Outrossim, retorna-se tamb√©m um valor iterativo
quando n√£o consumida.**

### **‚ö†Ô∏è Alerta para as fun√ß√µes "map()" e "filter()"‚ö†Ô∏è**
- _√â necess√°rio saber escolher entre o momento certo de usar "list comprehension" ou "map()" + "filter()"_

```python
list_of_prices = [100, 1000, 4250, 3020]

filtered_numbers = list(
    map(
        lambda filtered_value: filtered_value ** 2,
        filter(lambda value: value > 2000, list_of_prices)
    )
)
print(filtered_numbers)
```

```python
list_of_prices = [100, 1000, 4250, 3020]

filtered_numbers = [value ** 2 for value in list_of_prices if value > 2000]
print(filtered_numbers)
```

**Em casos como esse, usar list comprehension √© a melhor op√ß√£o. Pois √© mais simples de ler e menos complexo. Sendo assim, a forma mais "pythonica".**

## **_üìåNO FINAL DE TUDO, O QUE √â REALMENTE A FUN√á√ÉO "MAP()"?üìå_**

- **_A fun√ß√£o map() aplica uma fun√ß√£o a cada elemento de um iter√°vel (lista, tupla, etc.) e retorna um objeto iterador contendo os resultados dessa transforma√ß√£o._**
- **_N√£o cria listas automaticamente: para ver os resultados, voc√™ precisa converter o iterador (ex.: list(map(...)))._**
- **_Pode ser usada junto com lambda, filter() e reduce(), principalmente em um estilo mais funcional de programa√ß√£o._**
